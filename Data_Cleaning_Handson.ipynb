2. Correcting the Data Types

```python
#Checking the datatypes of all the columns
for i in df.columns:
  print("Column Name: ",i)
  print("Data Type: ", df[i].dtype)
  print()
```

**Insights**:
>1. The price and service fee columns are currently in the object data type, we need to convert them into numeric data type.
>2. The last review column is in object datatype, it should be converted into datatime data type.

```python
#Removing the dollar sign from the service fee column
#Removing the dollar sign from the service fee column

df['service fee'] = df['service fee'].str.replace("[^0-9]", '', regex=True).str.strip()
```

```python
#Converting the service fee column to numeric datatype

df['service fee'] = pd.to_numeric(df['service fee'])
```

```python
print(df['service fee'].dtype)
```

```python
#Removing the dollar sign from the price column

df['price'] = df['price'].str.replace("[^0-9]", '', regex=True).str.strip()
```

```python
#Converting the price column to numeric datatype

df['price'] = pd.to_numeric(df['price'])
```

```python
print(df['price'].dtype)
```

```python
#Converting the last review column to datetime datatype

df['last review'] = pd.to_datetime(df['last review'])
```

```python
print(df['last review'].dtype)
```

**Data Cleaning: Type Casting**:
> Converted both the service fee and price columns to numeric data type  
> Converted the last review column datatype to datetime

3. Handling the missing values

```python
#Checking for null values
df.isnull().sum()
```

```python
#Dropped the license column because it has almost 95 percent missing values and also it provides less information

df = df.drop(columns='license')
```

```python
df.head()
```

```python
df['service fee'] = df['service fee'].fillna(df['service fee'].median)
df['price'] = df['price'].fillna(df['price'].median)
```

```python
df['reviews per month'] = df['reviews per month'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['last review'] = df['last review'].fillna('No Review')
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['house_rules'] = df['house_rules'].fillna('Not Given')
```

```python
df = df.dropna()
```

```python
df.isnull().sum().sum()
```

**Data Cleaning**: **Handling Missing Values**:
> 1. Dropped the license column as it holds almost 99 percent missing values and it is also less informative.  
> 2. Replaced the missing values in both 'price' and 'service fee' columns with median of the respective columns.
> 3. Filled the missing values in the 'reviews per month' with (0).
> 4. Replaced the missing values in the 'last review' column with 'No Review'.
> 5. Replaced the missing values in the 'house_rules' column with 'Not Given'.
